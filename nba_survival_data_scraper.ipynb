```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
column_headers = ['Player', 'From', 'To', 'Pos', 'Ht', 'Wt', 'Birth Date', 'College']
```

```python
alphabet_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
player_data = []

for alphabet in alphabet_list:
    url_base = 'https://www.basketball-reference.com/players/' + alphabet
    result = requests.get(url_base)
    c = result.content
    soup = BeautifulSoup(c, 'lxml')
    data_rows = soup.findAll('tr')[1:]
    player_data_temp = [[td.getText() for td in data_rows[i].findAll(['th', 'td'])] 
                           for i in range(len(data_rows))]
    player_data += player_data_temp
```

```python
df = pd.DataFrame(player_data, columns=column_headers)
```

```python
df = df.convert_objects(convert_numeric=True)
df['Years_Played'] = df['To'] - df['From'] + 1
df['Censored'] = np.where(df['To'] == 2018, 0, 1)
df.head(10)
```
